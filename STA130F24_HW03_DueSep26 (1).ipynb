````python
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import numpy as np

# Load penguins dataset
penguins = sns.load_dataset("penguins").dropna()

# Create species-specific histograms
species = penguins['species'].unique()

# Set up a figure
fig = go.Figure()

# Add histograms for each species
for spec in species:
    # Subset data
    data = penguins[penguins['species'] == spec]['flipper_length_mm']
    
    # Add histogram for current species
    fig.add_trace(go.Histogram(x=data, name=spec, opacity=0.6))

    # Calculate statistics
    mean = np.mean(data)
    median = np.median(data)
    min_val = np.min(data)
    max_val = np.max(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    std = np.std(data)
    
    # Add horizontal lines (mean and median)
    fig.add_vline(x=mean, line=dict(color='blue', dash='dash'), annotation_text=f'{spec} Mean', annotation_position="top right")
    fig.add_vline(x=median, line=dict(color='green', dash='dot'), annotation_text=f'{spec} Median', annotation_position="bottom right")
    
    # Add horizontal rectangles (range, IQR, and 2 std deviations)
    fig.add_vrect(x0=min_val, x1=max_val, line_width=0, fillcolor="red", opacity=0.2, annotation_text=f'{spec} Range')
    fig.add_vrect(x0=q1, x1=q3, line_width=0, fillcolor="yellow", opacity=0.3, annotation_text=f'{spec} IQR')
    fig.add_vrect(x0=mean - 2 * std, x1=mean + 2 * std, line_width=0, fillcolor="blue", opacity=0.1, annotation_text=f'{spec} Â±2 Std')

# Update layout
fig.update_layout(barmode='overlay', title='Flipper Length Distribution with Statistical Markers',
                  xaxis_title='Flipper Length (mm)', yaxis_title='Count')

# Show plot
fig.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load penguins dataset
penguins = sns.load_dataset("penguins").dropna()

# Create species-specific KDE plots
species = penguins['species'].unique()

# Set up a figure with 3 subplots (one for each species)
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Loop over each species and create a KDE plot
for i, spec in enumerate(species):
    # Subset data for the current species
    data = penguins[penguins['species'] == spec]['flipper_length_mm']
    
    # Plot KDE for current species
    sns.kdeplot(data=data, ax=axes[i], fill=True, color='skyblue')
    axes[i].set_title(f'{spec} KDE Plot')
    axes[i].set_xlabel('Flipper Length (mm)')
    
    # Calculate statistics
    mean = np.mean(data)
    median = np.median(data)
    min_val = np.min(data)
    max_val = np.max(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    std = np.std(data)
    
    # Add vertical lines for mean and median
    axes[i].axvline(mean, color='blue', linestyle='--', label='Mean')
    axes[i].axvline(median, color='green', linestyle=':', label='Median')
    
    # Add shaded areas for range, IQR, and Â±2 standard deviations
    axes[i].axvspan(min_val, max_val, color='red', alpha=0.2, label='Range')
    axes[i].axvspan(q1, q3, color='yellow', alpha=0.3, label='IQR')
    axes[i].axvspan(mean - 2 * std, mean + 2 * std, color='blue', alpha=0.1, label='Â±2 Std Dev')
    
    # Add legend
    axes[i].legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
````

Right-Skewed Distribution:

The distribution has a long tail on the right (i.e., a few large values).
In right-skewed data, the mean is greater than the median because the mean is pulled towards the larger outliers in the right tail.
Example: The Gamma distribution used in the code demonstrates this, where the right tail causes the mean to be higher than the median.
Left-Skewed Distribution:

The distribution has a long tail on the left (i.e., a few small values).
In left-skewed data, the mean is less than the median because the mean is pulled towards the smaller outliers in the left tail.
Example: The negative of the Gamma distribution (mirroring it) shows left skewness, where the mean is lower than the median.
Cause of the Mean-Median Difference:

The mean is sensitive to extreme values (outliers) since it accounts for the magnitude of all data points.
The median, being the middle value, is less affected by extreme outliers.
As a result, in skewed distributions, the mean is shifted toward the tail, while the median stays closer to the majority of data.

This corresponds to more than half of the data being greater than the mean and the data being distributed closer to the right. A right-skewed distribution is farther from the right and the median is less than the mean. This corresponds to more than half of the data being less than the mean and the data being distributed closer to the left.


````python
import numpy as np
import pandas as pd
import plotly.express as px
from scipy import stats

sample_right_skew = stats.gamma(a=2, scale=2).rvs(size=1000)

mean_right = sample_right_skew.mean()
median_right = np.quantile(sample_right_skew, 0.5)

df_right_skew = pd.DataFrame({'data': sample_right_skew})

fig_right_skew = px.histogram(df_right_skew, x='data', nbins=50, title=f"Right-Skewed Distribution (Gamma): Mean = {mean_right:.2f}, Median = {median_right:.2f}")
fig_right_skew.add_vline(x=mean_right, line_width=2, line_dash="dash", line_color="blue", annotation_text="Mean", annotation_position="top left")
fig_right_skew.add_vline(x=median_right, line_width=2, line_dash="dash", line_color="green", annotation_text="Median", annotation_position="top right")

fig_right_skew.show(renderer="png")

sample_left_skew = -stats.gamma(a=2, scale=2).rvs(size=1000)

mean_left = sample_left_skew.mean()
median_left = np.quantile(sample_left_skew, 0.5)

df_left_skew = pd.DataFrame({'data': sample_left_skew})

fig_left_skew = px.histogram(df_left_skew, x='data', nbins=50, title=f"Left-Skewed Distribution (Negative Gamma): Mean = {mean_left:.2f}, Median = {median_left:.2f}")
fig_left_skew.add_vline(x=mean_left, line_width=2, line_dash="dash", line_color="blue", annotation_text="Mean", annotation_position="top left")
fig_left_skew.add_vline(x=median_left, line_width=2, line_dash="dash", line_color="green", annotation_text="Median", annotation_position="top right")

fig_left_skew.show(renderer="png")

sample_normal = np.random.normal(loc=0, scale=1, size=1000)

mean_normal = sample_normal.mean()
median_normal = np.quantile(sample_normal, 0.5)

df_normal = pd.DataFrame({'data': sample_normal})

fig_normal = px.histogram(df_normal, x='data', nbins=50, title=f"Normal Distribution: Mean = {mean_normal:.2f}, Median = {median_normal:.2f}")
fig_normal.add_vline(x=mean_normal, line_width=2, line_dash="dash", line_color="blue", annotation_text="Mean", annotation_position="top left")
fig_normal.add_vline(x=median_normal, line_width=2, line_dash="dash", line_color="green", annotation_text="Median", annotation_position="top right")

fig_normal.show(renderer="png")
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/manuelamc14/fast-food-Nutritional-Database/main/Tables/nutrition.csv")

# Display the first few rows
print(df.head())

# Check for missing values
print(df.isnull().sum())

# Get an overview of the data (types, non-null counts, etc.)
print(df.info())

# Summary statistics for all numeric columns
summary_stats = df.describe()

# Display summary statistics
print(summary_stats)

import numpy as np

# Simulating 100 days of returns for diversified and focused portfolios
np.random.seed(42)  # For reproducibility

# Simulate diversified portfolio (10 stocks, averaging their returns)
diversified_returns = np.mean(np.random.normal(0.05, 0.02, (100, 10)), axis=1)

# Simulate focused portfolio (3 high-potential stocks)
focused_returns = np.mean(np.random.normal(0.07, 0.05, (100, 3)), axis=1)

# Create a dataframe to store the results
portfolio_df = pd.DataFrame({
    'Day': np.arange(1, 101),
    'Diversified': diversified_returns,
    'Focused': focused_returns
})

# Cumulative returns over time
portfolio_df['Diversified_Cumulative'] = (1 + portfolio_df['Diversified']).cumprod()
portfolio_df['Focused_Cumulative'] = (1 + portfolio_df['Focused']).cumprod()

# Display the simulated data
print(portfolio_df.head())

# Plot cumulative returns
plt.figure(figsize=(10, 6))
plt.plot(portfolio_df['Day'], portfolio_df['Diversified_Cumulative'], label='Diversified Portfolio')
plt.plot(portfolio_df['Day'], portfolio_df['Focused_Cumulative'], label='Focused Portfolio')
plt.title('Cumulative Returns: Diversified vs Focused Portfolio')
plt.xlabel('Day')
plt.ylabel('Cumulative Return')
plt.legend()
plt.grid(True)
plt.show()

# Plot histogram of daily returns for both portfolios
plt.figure(figsize=(10, 6))
plt.hist(portfolio_df['Diversified'], bins=20, alpha=0.5, label='Diversified Portfolio')
plt.hist(portfolio_df['Focused'], bins=20, alpha=0.5, label='Focused Portfolio')
plt.title('Distribution of Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
````
